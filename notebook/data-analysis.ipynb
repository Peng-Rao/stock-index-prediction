```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
sp500 = yf.download("^GSPC", period="max", start="2001-1-1")
sp500.head()
```

```python
sp500["50ma"] = sp500["Close"].rolling(window=50).mean()
sp500["200ma"] = sp500["Close"].rolling(window=200).mean()
```

```python
plt.figure(figsize=(15, 5))
plt.title("S&P 500 Close Price with 200 and 50-Day Moving Average")
plt.plot(sp500["Close"])
plt.plot(sp500["50ma"])
plt.plot(sp500["200ma"])
plt.legend(["Close", "50MA", "200MA"])
plt.show()
```

```python
# Calculate the daily return of the close price
sp500["return"] = sp500["Close"].pct_change()

# Plot the daily return
plt.figure(figsize=(15, 5))
plt.title("S&P 500 Daily Return")
plt.plot(sp500["return"])
plt.show()
```

```python
# Plot the daily return using an histogram
plt.figure(figsize=(15, 3))
plt.title("S&P 500 Daily Return")
sp500["return"].hist(bins=200, grid=False)
plt.show()
```

```python
nsdk_data = pd.read_csv("../data/A纳斯达克综合指数.csv", index_col=0, parse_dates=True)
dqs_data = pd.read_csv("../data/A道琼斯工业平均指数.csv", index_col=0, parse_dates=True)
bzpr_data = pd.read_csv("../data/A标准普尔500指数.csv", index_col=0, parse_dates=True)
sh_data = pd.read_csv("../data/A上证指数.csv", index_col=0, parse_dates=True)
sz_data = pd.read_csv("../data/A深证指数.csv", index_col=0, parse_dates=True)
cyb_data = pd.read_csv("../data/A创业板指数.csv", index_col=0, parse_dates=True)
```

```python
stock_market = [0, nsdk_data, dqs_data, bzpr_data, sh_data, sz_data, cyb_data]
stock_name = [
    "",
    "Nasdaq Composite Index",
    "DowJones Industrial Average Index",
    "S&P 500 Index",
    "Shanghai Composite Index",
    "Shenzhen Index",
    "ChiNext Index",
]
stock_index = 1
data = stock_market[stock_index]


# Plot the line chart
plt.figure(figsize=(20, 6))
plt.plot(data.index, data["Close"], label="Index Values")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title(stock_name[stock_index] + " - Historical Close Prices")
plt.legend()
plt.grid()
plt.show()
```
