```python
stock_df = ak.stock_zh_index_spot_sina()
```

```python
stock_df = stock_df[["代码", "名称"]]
stock_df.rename(columns={"代码": "symbol", "名称": "name"}, inplace=True)
```

```python
stock_df.head()
```

```python
from sqlmodel import Field, SQLModel


class Stock(SQLModel, table=True):
    id: int | None = Field(default=None, primary_key=True)
    symbol: str
    name: str
```

```python
from typing import List

import pandas as pd


def df_to_sqlmodel(df: pd.DataFrame) -> List[SQLModel]:
    """Convert a pandas DataFrame into a a list of SQLModel objects."""
    objs = [Stock(**row) for row in df.to_dict("records")]
    return objs
```

```python
MYSQL_URL = "mysql+pymysql://root:123456@localhost:3306/A_stock?charset=utf8"
POOL_SIZE = 20
POOL_RECYCLE = 3600
POOL_TIMEOUT = 15
MAX_OVERFLOW = 2
CONNECT_TIMEOUT = 60
```

```python
from sqlmodel import create_engine

engine = create_engine(
    MYSQL_URL,
    pool_size=POOL_SIZE,
    pool_recycle=POOL_RECYCLE,
    pool_timeout=POOL_TIMEOUT,
    max_overflow=MAX_OVERFLOW,
    echo=True,
)
SQLModel.metadata.create_all(engine)
```

```python
from sqlmodel import Session

with Session(engine) as session:
    objs = df_to_sqlmodel(stock_df)
    session.add_all(objs)
    session.commit()
```
