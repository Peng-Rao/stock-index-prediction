```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x) / np.sum(np.exp(x), axis=0)


def RELU(x):
    """It returns zero if the input is less than zero otherwise it returns the given input."""
    x1 = []
    for i in x:
        if i < 0:
            x1.append(0)
        else:
            x1.append(i)

    return x1


def sigmoid(x):
    """It returns 1/(1+exp(-x)). where the values lies between zero and one"""

    return 1 / (1 + np.exp(-x))


def tanh(x):
    """It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1."""

    return np.tanh(x)


# 创建一个包含四个子图的画布
fig, axs = plt.subplots(1, 4, figsize=(20, 5))


x = np.linspace(-10, 10)
axs[0].plot(x, softmax(x))
axs[0].axis("tight")
axs[0].set_title("Activation Function :Softmax")
axs[1].plot(x, sigmoid(x))
axs[1].axis("tight")
axs[1].set_title("Activation Function :Sigmoid")
axs[2].plot(x, RELU(x))
axs[2].axis("tight")
axs[2].set_title("Activation Function :ReLU")
axs[3].plot(x, tanh(x))
axs[3].axis("tight")
axs[3].set_title("Activation Function :Tanh")
# 调整子图之间的间距
plt.tight_layout()
plt.show()
```

```python
# 保存图片，高分辨率，保存为 png 格式
fig.savefig("activation_functions.png", format="png")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一些示例数据
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x**2)

# 创建一个包含四个子图的画布
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

# 在每个子图中绘制数据
for i, ax in enumerate(axs):
    ax.plot(x, y, label=f"Plot {i+1}")
    ax.legend()

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```
